```python
# importando as bibliotecas para serem usadas
import pandas as pd
import SimpSOM as sps
from sklearn.cluster import KMeans
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# importando o dataset iris 
X, y = load_iris(return_X_y=True)
```

```python
# treinando a rede SOM no dataset iris
net = sps.somNet(20, 20, X, PBC=True)
net.train(0.01, 1000)
net.nodes_graph(colnum=0)
```

```python
# retornando os clusters criados
net.cluster(X)
```

```python
# projetando os dados em um gráfico 2D
prj = np.array(net.project(X))
plt.scatter(prj.T[0],prj.T[1])
plt.show()
```
