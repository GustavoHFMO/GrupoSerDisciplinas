```python
import numpy as np
from ypstruct import structure


class geneticAlgorithm:
    '''
    Classe para instanciar o algoritmo genetico
    '''
    pass

    def roulette_wheel_selection(self, p):
        '''
        Metodo para selecionar aleatoriamente um individuo
        :p: probabilidade de escolher um individuo
        '''
        c = np.cumsum(p)
        r = sum(p)*np.random.rand()
        ind = np.argwhere(r <= c)
        return ind[0][0]

    def apply_bound(self, x, varmin, varmax):
        '''
        Metodo para limitar a busca dos individuos
        :x: individuo
        :varmin: posicao minima
        :varmax: posicao maxima
        '''
        x.position = np.maximum(x.position, varmin)
        x.position = np.minimum(x.position, varmax)
    
    def mutate(self, x, mu, sigma):
        '''
        Metodo para aplicar a mutacao nos individuos
        :x: individuo 
        :mu: parametro para atribuir mutacao
        :sigma: parametro para alterar o gene
        '''
        y = x.deepcopy()
        flag = np.random.rand(*x.position.shape) <= mu
        ind = np.argwhere(flag)
        y.position[ind] += sigma*np.random.randn(*ind.shape)
        return y
    
    def crossover(self, p1, p2, gamma=0.1):
        '''
        Metodo para cruzar dois individuos
        :p1: individuo 1
        :p2: individuo 2
        :gamma: parametro para alterar o gene alterado
        '''
        
        c1 = p1.deepcopy()
        c2 = p1.deepcopy()
        alpha = np.random.uniform(-gamma, 1+gamma, *c1.position.shape)
        c1.position = alpha*p1.position + (1-alpha)*p2.position
        c2.position = alpha*p2.position + (1-alpha)*p1.position
        return c1, c2

    def run(self, problem, params):
        '''
        Metodo para executar o algoritmo genetico
        :problem: recebe a funcao a ser otimizada
        :params: recebe os parametros da funcao
        '''

        # Informacao do problema
        costfunc = problem.costfunc
        nvar = problem.nvar
        varmin = problem.varmin
        varmax = problem.varmax

        # parametros do algoritmo genetico
        maxit = params.maxit
        npop = params.npop
        beta = params.beta
        pc = params.pc
        nc = int(np.round(pc*npop/2)*2)
        gamma = params.gamma
        mu = params.mu
        sigma = params.sigma

        # Criando um individuo inicial
        empty_individual = structure()
        empty_individual.position = None
        empty_individual.cost = None

        # Melhor solucao ja encontrada
        bestsol = empty_individual.deepcopy()
        bestsol.cost = np.inf

        # Inicializando a populacao
        pop = empty_individual.repeat(npop)
        for i in range(npop):
            pop[i].position = np.random.uniform(varmin, varmax, nvar)
            pop[i].cost = costfunc(pop[i].position)
            if pop[i].cost < bestsol.cost:
                bestsol = pop[i].deepcopy()

        # Melhor fitness ja encontrado
        bestcost = np.empty(maxit)

        # Repetir o processo do GA
        for it in range(maxit):

            costs = np.array([x.cost for x in pop])
            avg_cost = np.mean(costs)
            if avg_cost != 0:
                costs = costs/avg_cost
            probs = np.exp(-beta*costs)

            popc = []
            for _ in range(nc//2):


                # realizado a selecao dos individuos
                p1 = pop[roulette_wheel_selection(probs)]
                p2 = pop[roulette_wheel_selection(probs)]

                # realizando o crossover
                c1, c2 = crossover(p1, p2, gamma)

                # realizando a mutacao
                c1 = mutate(c1, mu, sigma)
                c2 = mutate(c2, mu, sigma)

                # restringindo a busca
                apply_bound(c1, varmin, varmax)
                apply_bound(c2, varmin, varmax)

                # encontrando o melhor fitness
                c1.cost = costfunc(c1.position)
                if c1.cost < bestsol.cost:
                    bestsol = c1.deepcopy()

                # encontrando o segundo melhor fitness
                c2.cost = costfunc(c2.position)
                if c2.cost < bestsol.cost:
                    bestsol = c2.deepcopy()

                # adicionando os descendentes
                popc.append(c1)
                popc.append(c2)


            pop += popc
            pop = sorted(pop, key=lambda x: x.cost)
            pop = pop[0:npop]

            # guardando o melhor fitness
            bestcost[it] = bestsol.cost

            # mostrando a iteracao atual
            print("Iteration {}: Best Cost = {}".format(it, bestcost[it]))

        # saida do metodo
        out = structure()
        out.pop = pop
        out.bestsol = bestsol
        out.bestcost = bestcost
        return out
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ypstruct import structure

# Sphere Test Function
def sphere(x):
    return sum(x**2)

# Problem Definition
problem = structure()
problem.costfunc = sphere
problem.nvar = 5
problem.varmin = [-10, -10, -1, -5,  4]
problem.varmax = [ 10,  10,  1,  5, 10]

# GA Parameters
params = structure()
params.maxit = 100
params.npop = 50
params.beta = 1
params.pc = 1
params.gamma = 0.1
params.mu = 0.01
params.sigma = 0.1

# Run GA
ga = geneticAlgorithm()
out = ga.run(problem, params)

# Results
plt.plot(out.bestcost)
# plt.semilogy(out.bestcost)
plt.xlim(0, params.maxit)
plt.xlabel('Iterations')
plt.ylabel('Best Cost')
plt.title('Genetic Algorithm (GA)')
plt.grid(True)
plt.show()
```
